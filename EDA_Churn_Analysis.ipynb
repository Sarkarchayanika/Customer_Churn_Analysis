```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('Customer Churn.csv')
df.head()
```

```python
df.info()
```

#replacing the blank as 0 as tenure is 0 and no total charges are recorded

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ", "0")
df["TotalCharges"] =df["TotalCharges"].astype("float")
```

#creating a function for the conversion of Senior Citizen Value from 0/1 to Yes/No

```python
def conversion(value):
    if value == 1:
        return "Yes"
    else:
        return "No"

df["SeniorCitizen"] = df["SeniorCitizen"].apply(conversion)
```

```python
df.head(5)
```

#creating a bar chart to show the value of churn out customers

```python
ax = sns.countplot(x = 'Churn', data=df)
ax.bar_label(ax.containers[0])
plt.show()
```

```python
plt.figure(figsize = (3,4))

gb = df.groupby("Churn").agg({'Churn' : 'count'})
plt.title("Count of customer that has been churned out")
plt.pie(gb['Churn'] , labels = gb.index, autopct = "%1.2f%%")
plt.show
```

#from the above analysis we can see that 26.54% of customers have churned out

```python
plt.figure(figsize = (4,4))
ax = sns.countplot(x = "SeniorCitizen", data=df)
ax.bar_label (ax.containers[0])
plt.title("Churn by Senior Citizen", fontsize = 8)
plt.show()
```

```python
count_data = df.groupby(['SeniorCitizen', 'Churn']).size().unstack(fill_value=0)

# Calculate percentage relative to the total count per SeniorCitizen
count_data_percentage = count_data.divide(count_data.sum(axis=1), axis=0) * 100

# Plot stacked bar chart
plt.figure(figsize=(6, 4))
ax = count_data_percentage.plot(kind='bar', stacked=True, color=['skyblue', 'lightcoral'])

# Add title and labels
plt.title('Churn by SeniorCitizen (%)', fontsize=12)
plt.xlabel('SeniorCitizen', fontsize=10)
plt.ylabel('Percentage', fontsize=10)

# Add percentage labels on top of each bar
for p in ax.patches:
    height = p.get_height()
    width = p.get_width()
    x = p.get_x() + width / 2
    y = p.get_y() + height / 2
    ax.text(x, y, f'{height:.1f}%', ha='center', va='center', fontsize=8, color='black')

plt.show()
```

```python
# List of service-related columns
services = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 
            'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

# Set the figure size for the entire grid
plt.figure(figsize=(15, 12))

# Loop through each service and create a count plot for it
for i, service in enumerate(services, 1):
    plt.subplot(3, 3, i)  # 3x3 grid of subplots (you can adjust the grid size based on your preference)
    sns.countplot(x=service, data=df, hue= df["Churn"])
    plt.title(f"Distribution of {service}")
    plt.xticks(rotation=45)  # Rotate labels for better readability

# Adjust layout to make sure plots don't overlap
plt.tight_layout()
plt.show()
```
